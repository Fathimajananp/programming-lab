```python
import rectangle
import circle

while(1):
    print("\n1.RECTANGLE\n2.CIRCLE\n3.EXIT\n")
    ch=int(input("Enter your choice:"))
    if ch==1:
        l=int(input("Enter the length of the rectangle:"))
        b=int(input("Enter the breadth of the rectangle:"))
        rectangle.recArea(l,b)
        rectangle.recPerimeter(l,b)
    elif ch==2:
        r=int(input("Enter the radius of the circle:"))
        circle.circleArea(r)
        circle.circlePerimeter(r)
    else:
        print("Exiting...")
        break
```

```python
from Graphics import rectangle
from Graphics import circle
from Graphics.Graphics3D import cuboid
from Graphics.Graphics3D import sphere
while(1):
    print("\n1.RECTANGLE\n2.CIRCLE\n3.CUBOID\n4.SPHERE\n5.EXIT\n")
    ch=int(input("Enter your choice :"))
    if ch==1:
        l=int(input("Enter the length of the rectangle:"))
        b=int(input("Enter the breadth of the rectangle:"))
        rectangle.recArea(l,b)
        rectangle.recPerimeter(l,b)
    elif ch==2:
        r=int(input("Enter the radius of the circle:"))
        circle.circleArea(r)
        circle.circlePerimeter(r)
    elif ch==3:
        l=int(input("Enter the length of the cuboid:"))
        b=int(input("Enter the breadth of the cuboid:"))
        h=int(input("Enter the height of the cuboid:"))
        cuboid.cuboidArea(l,b,h)
        cuboid.cuboidPerimeter(l,b,h)
    elif ch==4:
        r=int(input("Enter the radius of the sphere:"))
        sphere.sphereArea(r)
        sphere.sphereCircumference(r)
    else:
        print("Exiting...")
        break
        
```
